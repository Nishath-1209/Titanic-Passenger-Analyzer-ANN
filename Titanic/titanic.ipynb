```python
import pandas as pd
data = pd.read_csv("Titanic-Dataset.csv")

print(data.head())
```

# INPUT VALUES (Normalized)

```python
x1 = 0.20   # Pclass
x2 = 0.24   # Age
x3 = 0.80   # Fare

target = 1
# LEARNING RATE
lr = 0.1
```

```python
# INITIAL WEIGHTS
# Input -> Hidden
w1 = 0.11
w2 = 0.14
w3 = 0.17

w4 = 0.21
w5 = 0.24
w6 = 0.27

# Hidden Biases
bh1 = 0.1
bh2 = 0.1

# Hidden -> Output
w7 = 0.31
w8 = 0.34

# Output Bias
bo = 0.1
```

```python
# SIGMOID FUNCTION
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
# FORWARD PROPAGATION

print("===== FORWARD PROPAGATION =====")

# Hidden neuron h1
h1_net = (x1 * w1) + (x2 * w2) + (x3 * w3) + bh1
h1 = sigmoid(h1_net)

print(f"h1 net input = {h1_net:.4f}")
print(f"h1 activation = {h1:.4f}")

# Hidden neuron h2
h2_net = (x1 * w4) + (x2 * w5) + (x3 * w6) + bh2
h2 = sigmoid(h2_net)

print(f"\nh2 net input = {h2_net:.4f}")
print(f"h2 activation = {h2:.4f}")

# Output neuron
o1_net = (h1 * w7) + (h2 * w8) + bo
output = sigmoid(o1_net)

print(f"\nOutput net input = {o1_net:.4f}")
print(f"Predicted Output = {output:.4f}")

# ERROR CALCULATION

print("\n===== ERROR CALCULATION =====")

error = 0.5 * ((target - output) ** 2)

print(f"Mean Squared Error = {error:.4f}")
```

```python
# BACKPROPAGATION
print("\n===== BACKPROPAGATION =====")
# Output layer gradient
delta_output = (output - target) * output * (1 - output)
print(f"Output Gradient = {delta_output:.4f}")
# Hidden layer gradients
delta_h1 = h1 * (1 - h1) * (delta_output * w7)
delta_h2 = h2 * (1 - h2) * (delta_output * w8)
print(f"Hidden Gradient h1 = {delta_h1:.4f}")
print(f"Hidden Gradient h2 = {delta_h2:.4f}")
# =========================
# UPDATE HIDDEN -> OUTPUT

print("\n===== UPDATE HIDDEN → OUTPUT WEIGHTS =====")

w7_new = w7 - (lr * delta_output * h1)
w8_new = w8 - (lr * delta_output * h2)

bo_new = bo - (lr * delta_output)

print(f"Updated w7 = {w7_new:.4f}")
print(f"Updated w8 = {w8_new:.4f}")
print(f"Updated bo = {bo_new:.4f}")

# =========================
# UPDATE INPUT -> HIDDEN
# =========================

print("\n===== UPDATE INPUT → HIDDEN WEIGHTS =====")

w1_new = w1 - (lr * delta_h1 * x1)
w2_new = w2 - (lr * delta_h1 * x2)
w3_new = w3 - (lr * delta_h1 * x3)

w4_new = w4 - (lr * delta_h2 * x1)
w5_new = w5 - (lr * delta_h2 * x2)
w6_new = w6 - (lr * delta_h2 * x3)

bh1_new = bh1 - (lr * delta_h1)
bh2_new = bh2 - (lr * delta_h2)

print(f"Updated w1 = {w1_new:.4f}")
print(f"Updated w2 = {w2_new:.4f}")
print(f"Updated w3 = {w3_new:.4f}")

print(f"\nUpdated w4 = {w4_new:.4f}")
print(f"Updated w5 = {w5_new:.4f}")
print(f"Updated w6 = {w6_new:.4f}")

print(f"\nUpdated bh1 = {bh1_new:.4f}")
print(f"Updated bh2 = {bh2_new:.4f}")
```

```python
print("\n===== FINAL UPDATED PARAMETERS =====")

print(f"w1 = {w1_new:.4f}")
print(f"w2 = {w2_new:.4f}")
print(f"w3 = {w3_new:.4f}")

print(f"w4 = {w4_new:.4f}")
print(f"w5 = {w5_new:.4f}")
print(f"w6 = {w6_new:.4f}")

print(f"w7 = {w7_new:.4f}")
print(f"w8 = {w8_new:.4f}")

print(f"bh1 = {bh1_new:.4f}")
print(f"bh2 = {bh2_new:.4f}")

print(f"bo = {bo_new:.4f}")
```
